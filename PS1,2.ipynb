## 1.Perform the following operations using Python on the Facebookmetrics datasets.
    a.Create data subsets
    b.Merge Data
    c.Transposing Data

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dataset_Facebook.csv', sep = ';')
```

```python
df.sample(4)
```

## a. Creating the subsets

```python
subset1 = df[df['Post Hour']<5]
```

```python
subset1.sample(3)
```

```python
subset2 = df[df['Type']=='Photo']
```

```python
subset2.sample(3)
```

## b. Merging Data

```python
merged_data = pd.merge(subset1,subset2)
```

```python
merged_data.sample(10) ## Merge data is a dataframe which contains post type = photos and post hour < 10.
```

## c. Transposing Data

```python
merged_data.T
```
