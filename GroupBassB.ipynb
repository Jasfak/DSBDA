```python
## Calculating mean/median values so that they can fill the NAN values present in these features
mean_PM2 = df ['PM2.5'].mean()
mean_PM1 = df ['PM10'].mean()
mean_NO = df ['NO'].mean()
mean_NO2 = df ['NO2'].mean()
mean_NOx = df ['NOx'].mean()
mean_NH3 = df ['NH3'].mean()
mean_CO = df ['CO'].mean()
mean_SO2 = df ['SO2'].mean()
mean_O3 = df ['O3'].mean()
mean_Benzene = df ['Benzene'].mean()
mean_Toluine = df ['Toluene'].mean()
mean_Xylene = df ['Xylene'].mean()
median_AQI = df ['AQI'].median()
```

```python
# Filling the calculated mean/ median values inplace of NaN values permanently by using 'inplace=True'
df ['PM2.5'].fillna(mean_PM2, inplace = True)
df ['PM10'].fillna(mean_PM1, inplace = True)
df ['NO'].fillna(mean_NO, inplace = True)
df ['NO2'].fillna(mean_NO2, inplace = True)
df ['NOx'].fillna(mean_NOx, inplace = True)
df ['NH3'].fillna(mean_NH3, inplace = True)
df ['CO'].fillna(mean_CO, inplace = True)
df ['SO2'].fillna(mean_SO2, inplace = True)
df ['O3'].fillna(mean_O3, inplace = True)
df ['Benzene'].fillna(mean_Benzene, inplace = True)
df ['Toluene'].fillna(mean_Toluine, inplace = True)
df ['Xylene'].fillna(mean_Xylene, inplace = True)
df ['AQI'].fillna(median_AQI, inplace = True)
```

```python
y  = df1['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25, random_state=0)
X_train.shape, X_test.shape
```

```python
from sklearn.svm import SVC   # Support vector classifier
```

```python
svc = SVC(kernel='linear')
```

```python
svc.fit(X_train,y_train)
```

```python
y_pred = svc.predict(X_test) #Prediction Variabler
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score 
```

```python
#Prediction Report :

print("CR:",classification_report(y_test,y_pred))
print('Confusion Matrix:\n\n',confusion_matrix(y_test,y_pred))
print('\nAccuracy Score :',accuracy_score(y_test,y_pred))
```
