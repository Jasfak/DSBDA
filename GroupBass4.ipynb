```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Air-Quality.csv')
```

```python
df.sample(3)
```

```python
plt.scatter(data = df, x = 'Ozone', y='Temp')
plt.xlabel('Ozone')
plt.title('Scatter plot of Ozone vs Temp')
plt.show()
```

```python
sns.histplot(x = 'Month', data = df)
```

```python
sns.boxplot(data=df, x = 'Wind')
```

```python
sns.heatmap(df.corr(),annot = True)
```

```python
monthly_means = df.groupby('Month')['Ozone'].mean()
plt.bar(monthly_means.index, monthly_means)
```

```python
sns.histplot(data=df, x='Temp', bins = 20, kde = True)
plt.show()
```

```python
sns.distplot(df['Temp'].dropna(), kde = True)
plt.show()
```
