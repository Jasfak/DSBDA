```python
import pandas as pd
df = pd.read_csv('dataset_Facebook.csv', sep=';') # seaparator is used to separate the csv file records
```

```python
df.sample(3)
```

```python
# Creating Subset 1

df_subset1=df.loc[0:4, ['like', 'share']] 
print('\n-------------------------- df_subset1 --------------------------\n')
print(df_subset1)
```

```python
# creating subset 2 just like subset 1 but in a different way

df_subset2=df.iloc[[0, 1, 2, 3, 4], [16, 17]]
print('\n-------------------------- df_subset2 --------------------------\n')
print(df_subset2)
```

```python
# creating subset 3 by calculating the like-share ration by using 'div' function

df_subset3=df['like'].div(df['share']).to_frame('like-share ratio')
print('\n-------------------------- df_subset3 --------------------------\n')
print(df_subset3)
```

```python
# Merging the two dataframes using concat() function

df_merged=pd.concat([df, df_subset3], axis=1, join='inner')
print('\n-------------------------- df_merged --------------------------\n')
print(df_merged)
```

```python
# Sorting rows according to Page total likes, Type and then like

df_sorted=df.sort_values(by=['Page total likes', 'Type', 'like'], ascending=False)
print('\n-------------------------- df_sorted --------------------------\n')
print(df_sorted)
```

```python
# D. TRANSPOSING DATA
# Creating a transpose of the dataframe using transpose() function

df_transposed=df.T # we can transpose the data either by using 'T' or 'transpose()' function
print('\n-------------------------- df_transposed --------------------------\n')
print(df_transposed)
```

```python
# E. SHAPE AND RESHAPE DATA
# Shape of the dataframe

df_shape=df.shape
print('\n-------------------------- df_shape --------------------------\n')
print(df_shape)
```

```python
# Reshape the dataframe using pivot() function 
# Shows the relationship between the Type of Post and the No. of likes

df_reshape=df.pivot(columns='Type', values='like') 
print('\n-------------------------- df_reshape --------------------------\n')
print(df_reshape)
```

### The pivot() method is used to reshape a DataFrame from a long to wide format. It is useful to get a DataFrame where one or more columns are identifier variables, and the other columns are pivoted to the column axis.
